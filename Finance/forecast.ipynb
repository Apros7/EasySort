```python
import matplotlib.pyplot as plt
```

### Production and CO2 Saved

```python
år = [0, 1]
ton_affald_sorteret_årligt = [0, 5]

for i in range(år[-1], 15):
    år.append(i)
    print(ton_affald_sorteret_årligt[-1]*2)
    ton_affald_sorteret_årligt.append(ton_affald_sorteret_årligt[-1]*2)

indkomst = [v*10000/1000000 for v in ton_affald_sorteret_årligt]
```

```python
plt.figure(figsize=(12, 4))

plt.subplot(1, 2, 1)
plt.plot(år, ton_affald_sorteret_årligt, marker='o', color='blue')
plt.title('Ton Affald Sorteret Årligt')
plt.xlabel('År')
plt.ylabel('Affald (Ton)')

plt.subplot(1, 2, 2)
plt.plot(år, indkomst, marker='o', color='green')
plt.title('Indkomst')
plt.xlabel('År')
plt.ylabel('Indkomst (mil. DKK)')

plt.tight_layout()
plt.show()
```
